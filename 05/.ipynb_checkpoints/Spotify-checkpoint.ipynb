```python
#1 print a list of 50 artist playable in the US - name and popularity

import requests

# query = 'lil' type = 'artist' number of responses(limit) = '50' country = 'US'
response = requests.get('https://api.spotify.com/v1/search?q=lil&type=artist&limit=50&country=US')

# data is a string, use json
data = response.json()

print(type(data))
print(data.keys())
print(type(data['artists']))
print(data['artists'].keys())
artists = data['artists']['items']
print(type(artists))
print(artists[0:3])

for artist in artists: 
    print(artist['name'], artist['popularity'])

for artist in artists: 
    print(artist['name'], artist['popularity'], artist['genres'])
    if artist['genres'] == []: 
        print ("No genres listed")
    
```
